**Példa:** Tekintsük a síkon a $(0,0), (1, 0), (0, 1)$ pontok által határolt zárt háromszöglapot, melyet jelöljünk $T$-vel.

**Kvadratúra rendje:** Mint (1,1) függvények estén. Azt mondjuk, hogy egy kvadratúra rendje $s$, ha minden legfeljebb $(s-1)$-edfokú polinomot pontosan integrál, azonban van olyan $s$-edfokú, amit már nem.

**Következmény:** Elég $x^ky^j$ alakú polinomokra nézni, ahol $(k+j \leq s)$, hiszen a kétváltozós, legfeljebb $s$-edfokú polinomok ilyeneknek a lineáris kombinációja.

Foglalkozzunk az 
$$
\int\limits_T f = \int\limits_0^1\int\limits_0^{1-x} f(x, y) \, dy \, dx
$$
integrál numerikus közelítésével.

### 1. Feladat

Mutassuk meg, hogy a
$$
\int\limits_{T} f \approx \frac12 f(1/3, 1/3)
$$
kvadratúra rendje

*a)* legalább 2, azaz pontos a legfeljebb elsőfokú polinomokon.

*b)* legfeljebb 2, azaz pontos a legfeljebb másodfokú polinomokon.

### 2. Feladat

$$ 
\int\limits_{T} f \approx \frac12 \frac{f(0, 1/2) + f(1/2, 1/2) + f(1/2, 0)}{3}
$$
kvadratúra rendje

*a)* legalább 3, azaz pontos a legfeljebb másodfokú polinomokon.

*b)* legfeljebb $3$, azaz van olyan harmadfokú polinom, amin már nem pontos.

Bonyolultabb alakzatok esetén háromszögelhetünk (nem csak háromszögekkel), azaz pl. az egységnégyzetet felbonthatjuk két olyan háromszögre, amin az integrálást már el tudjuk végezni.

### Egyszerű időintegrátorok

Bevezetésként tekintsük az

$$ \dot x(t) = \lambda \cdot x(t) $$

egyenletet, ahol $\lambda$ legyen egy valós szám. Ha a kezdeti feltételünk $x(0) = 1$, akkor a megoldás

$$x(t) = e^{\lambda t} = e^z = \sum_{j=0}^{\infty} \frac{z^j}{j!},$$
ahol $z = \lambda t$.
